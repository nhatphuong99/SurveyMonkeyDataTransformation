# Import Libraries

```python
import pandas as pd
import numpy as np
import os
```

# Get Directory Path

```python
pwd = os.getcwd()
pwd
```

# Import Question & Answer Data

```python
dataset = pd.read_excel(pwd + "/Data - Survey Monkey Output Edited.xlsx", sheet_name = "Edited_Data")
dataset
```

```python
dataset_modified = dataset.copy()
dataset_modified
```

# Drop Unnecessary Columns

```python
dataset_modified.columns
```

```python
columns_to_drop = ['Start Date', 'End Date', 'Email Address', 'First Name', 'Last Name', 'Custom Data 1']
columns_to_drop
```

```python
dataset_modified = dataset_modified.drop(columns = columns_to_drop)
dataset_modified
```

# Unpivot Data

```python
id_vars = list(dataset_modified.columns)[:8]
id_vars
```

```python
value_vars = list(dataset_modified.columns)[8:]
value_vars
```

```python
dataset_melted = dataset_modified.melt(id_vars = id_vars, value_vars = value_vars, var_name = "Question + Subquestion", value_name = "Answer")
dataset_melted
```

# Import Question Data

```python
questions_import = pd.read_excel(pwd + "/Data - Survey Monkey Output Edited.xlsx", sheet_name = "Question")
questions_import
```

```python
questions = questions_import.copy()
questions
```

```python
questions.columns
```

```python
questions = questions.drop(columns = ['Raw Question', 'Raw Subquestion', 'Subquestion', 'Unnamed: 5'])
questions
```

# Joint Question Data and Melted Data

```python
dataset_merged = pd.merge(left = dataset_melted, right = questions, how = "left", left_on = "Question + Subquestion", right_on = "Question + Subquestion")
print("Original Data: ", len(dataset_melted))
print("Merged Data: ", len(dataset_merged))
dataset_merged
```

# Find Number of Respondents

```python
respondents = dataset_merged[dataset_merged["Answer"].notna()]
respondents
```

```python
respondents_grouped = respondents.groupby("Question")["Respondent ID"].nunique().reset_index()
respondents_grouped.rename(columns = {"Respondent ID": "Number of Respondents"}, inplace = 1)
respondents_grouped
```

# Joint Number of Respondents to Merged Data

```python
dataset_merged_two = pd.merge(left = dataset_merged, right = respondents_grouped, how = "left", left_on = "Question", right_on = "Question")
print("Original Data: ", len(dataset_merged))
print("Merged Data: ", len(dataset_merged_two))
dataset_merged_two
```

# Find Number of Same Answers

```python
same_answer = dataset_merged#[dataset_merged["Answer"].notna()]
same_answer
```

```python
same_answer_grouped = same_answer.groupby(["Question + Subquestion", "Answer"])["Respondent ID"].nunique().reset_index()
same_answer_grouped.rename(columns = {"Respondent ID": "Number of Same Answers"}, inplace = 1)
same_answer_grouped
```

# Joint Number of Same Answers to Merged Data

```python
dataset_merged_three = pd.merge(left = dataset_merged_two, right = same_answer_grouped, how = "left", left_on = ["Question + Subquestion", "Answer"], right_on = ["Question + Subquestion", "Answer"])
print("Original Data: ", len(dataset_merged_two))
print("Merged Data: ", len(dataset_merged_three))
dataset_merged_three
```

# Fill NaN with 0

```python
dataset_merged_three["Number of Same Answers"].fillna(0, inplace = True)
dataset_merged_three
```

# Rename Columns

```python
columns_to_rename = {"Identify which division you work in. - Response": "Division Primary",
 "Identify which division you work in. - Other (please specify)": "Division Secondary",
 "Which of the following best describes your position level? - Response": "Position",
 "Which generation are you apart of? - Response": "Generation",
 "Please select the gender in which you identify. - Response": "Gender",
 "Which duration range best aligns with your tenure at your company? - Response": "Tenure",
 "Which of the following best describes your employment type? - Response": "Employment Type"}
```

```python
output = dataset_merged_three.copy()
output.rename(columns = columns_to_rename, inplace = True)
output
```

# Export Output

```python
output.to_excel(pwd + "/Final_Output.xlsx", index = False)
```
